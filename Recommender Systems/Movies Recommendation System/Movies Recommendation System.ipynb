#### Importing the Libraries

```python
import pandas as pd
import numpy as np
```

#### Importing the Dataset

```python
movies = pd.read_csv('imdb_top_1000.csv')
movies.head()
```

#### Filtering out necessary attributes necessary in Movie Recommendation

```python
# Series_Title 
# Released_Year 
# Certificate
# Genre 
# Overview
# Director
# Star1
# Star2
# Star3
# Star4
```

```python
movies = movies[['Series_Title', 'Released_Year', 'Certificate', 'Genre', 'Overview', 'Director', 'Star1', 'Star2', 'Star3', 'Star4']]
movies.head()
```

#### Transforming necessary columns values to list of single worded strings

```python
movies['Genre'] = movies['Genre'].apply(lambda x:x.split())
movies['Overview'] = movies['Overview'].apply(lambda x:x.split())
movies['Director'] = movies['Director'].apply(lambda x:x.split())
movies['Star1'] = movies['Star1'].apply(lambda x:x.split())
movies['Star2'] = movies['Star2'].apply(lambda x:x.split())
movies['Star3'] = movies['Star3'].apply(lambda x:x.split())
movies['Star4'] = movies['Star4'].apply(lambda x:x.split())
movies.head()
```

#### Removing spaces if any 

```python
# Define a function called "collapse" that takes a list "L" as an argument
def collapse(L):
    # Create an empty list called "L1"
    L1 = []
    # Iterate through each element in the input list "L"
    for i in L:
        # Remove any whitespace characters (i.e. spaces) from the current element using the "replace()" method with an empty string as the second argument
        # Then append the modified element to the "L1" list
        L1.append(i.replace(" ",""))
    # Return the modified list "L1"
    return L1
```

```python
movies['Genre'] = movies['Genre'].apply(collapse)
movies['Director'] = movies['Director'].apply(collapse)
movies['Overview'] = movies['Overview'].apply(collapse)
```

```python
movies.head()
```

#### Combining all the necessary columns that will help in Movie Recommendation

```python
movies['recommender'] = movies['Genre'] + movies['Overview'] + movies['Director'] + movies['Star1'] + movies['Star2'] + movies['Star3'] + movies['Star4']
movies.head()
```

```python
new_movies = movies.drop(labels = ['Genre', 'Overview', 'Director', 'Star1', 'Star2', 'Star3', 'Star4', 'Released_Year', 'Certificate'], axis = 1)
new_movies.head()
```

#### Converting list again to string

```python
new_movies['recommender'] = new_movies['recommender'].apply(lambda x:" ".join(x))
new_movies['recommender'] = new_movies['recommender'].apply(lambda x:x.lower())
new_movies.head()
```

#### Stemming words that have almost similar meaning

```python
from nltk.stem.porter import PorterStemmer
stemmer = PorterStemmer()
```

```python
def stem(text):
    L = []
    for i in text.split():
        L.append(stemmer.stem(i))
        
    return " ".join(L)
```

```python
new_movies['recommender'] = new_movies['recommender'].apply(stem)
```

```python
new_movies['recommender'][34]
```

#### Fetching top most occuring 3000 words in recommender column and vectorising them

```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(max_features = 3000, stop_words='english', )
```

```python
vectors = vectorizer.fit_transform(new_movies['recommender']).toarray()
```

```python
vectors.shape
```

```python
vectorizer.get_feature_names()
```

#### Calculating distance of each word with that from every word and forming a matrix for the same

```python
# Import the "cosine_similarity" function from the "pairwise" module in the "metrics" package of the "sklearn" library
from sklearn.metrics.pairwise import cosine_similarity
# Compute the cosine similarity matrix of the input matrix "vectors" using the "cosine_similarity" function
similarity = cosine_similarity(vectors)
```

#### Function to print the similar movies

```python
# Define a function called "recommend" that takes a movie title as an argument
def recommend(movie):
     # Get the index of the first matching movie in the DataFrame
    movie_index = movies[movies['Series_Title'] == movie].index[0]
    # Get the row of cosine similarity values between the input movie and all other movies in the "similarity" matrix
    distance = similarity[movie_index]
    # Sort the cosine similarity values in descending order and get the indices of the six most similar movies (excluding the input movie itself)
    movies_list = sorted(list(enumerate(distance)), reverse=True, key=lambda x:x[1])[1:7]
    # Loop through the indices of the six most similar movies and print their titles
    for i in movies_list:
        print(new_movies.iloc[i[0]].Series_Title)
```

```python
recommend('The Dark Knight')
```
